```python
y = list(train['TARGET'])
```

```python
train = train.drop('TARGET', axis=1)
```

```python
train.info()
```

```python
train.info(verbose=True, memory_usage=True, null_counts=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
from ipywidgets import Image
from io import StringIO
import pydotplus
from sklearn.tree import export_graphviz

dot_data = StringIO()

export_graphviz(dtc, feature_names=train.columns, out_file=dot_data, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(value=graph.create_png())
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
dtc_1 = DecisionTreeClassifier(max_depth=2)
dtc_1.fit(X_train, y_train)

dtc_2 = DecisionTreeClassifier(max_depth=5)
dtc_2.fit(X_train, y_train)

dtc_3 = DecisionTreeClassifier(max_depth=9)
dtc_3.fit(X_train, y_train)

dtc_4 = DecisionTreeClassifier(max_depth=15)
dtc_4.fit(X_train, y_train)
```

# DecisionTreeClassifier

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score

plt.figure(figsize=(10, 10))

predict_DTC_1 = dtc_1.predict_proba(X_test)
predict_DTC_2 = dtc_2.predict_proba(X_test)
predict_DTC_3 = dtc_3.predict_proba(X_test)
predict_DTC_4 = dtc_4.predict_proba(X_test)


fpr, tpr, thresholds = roc_curve(y_test, predict_DTC_1[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('DecisionTree_1',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_DTC_2[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('DecisionTree_2',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_DTC_3[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('DecisionTree_3',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_DTC_4[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('DecisionTree_4',auc(fpr, tpr)))



plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.legend(loc=0, fontsize='small')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

# LogisticRegression

```python
# %matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score




lr_1 = LogisticRegression(C=0.01, penalty='l1')
lr_1.fit(X_train, y_train)

lr_2 = LogisticRegression(C=0.01, penalty='l2')
lr_2.fit(X_train, y_train)

lr_3 = LogisticRegression(C=1, penalty='l1')
lr_3.fit(X_train, y_train)

lr_4 = LogisticRegression(C=1, penalty='l2')
lr_4.fit(X_train, y_train)

lr_5 = LogisticRegression(C=100, penalty='l1')
lr_5.fit(X_train, y_train)

lr_6 = LogisticRegression(C=100, penalty='l2')
lr_6.fit(X_train, y_train)


plt.figure(figsize=(20, 20))

predict_LR_1 = lr_1.predict_proba(X_test)
predict_LR_2 = lr_2.predict_proba(X_test)
predict_LR_3 = lr_3.predict_proba(X_test)
predict_LR_4 = lr_4.predict_proba(X_test)
predict_LR_5 = lr_5.predict_proba(X_test)
predict_LR_6 = lr_6.predict_proba(X_test)

fpr, tpr, thresholds = roc_curve(y_test, predict_LR_1[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('LogisticRegression_1',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_LR_2[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('LogisticRegression_2',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_LR_3[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('LogisticRegression_3',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_LR_4[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('LogisticRegression_4',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_LR_5[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('LogisticRegression_5',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_LR_6[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('LogisticRegression_6',auc(fpr, tpr)))

plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.legend(loc=0, fontsize='small')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
import numpy as np
from sklearn import ensemble
rf = ensemble.RandomForestClassifier( random_state=11)
rf.fit(X_train, y_train)



importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names=train.columns
print("Feature importances:")
for f, idx in enumerate(indices):
    print("{:2d}. feature '{:5s}' ({:.4f})".format(f + 1, feature_names[idx], importances[idx]))
```

```python
d_first = 30
plt.figure(figsize=(20, 20))
plt.title("Feature importances")
plt.bar(range(d_first), importances[indices[:d_first]], align='center')
plt.xticks(range(d_first), np.array(feature_names)[indices[:d_first]], rotation=90)
plt.xlim([-1, d_first]);
```

```python
best_features = indices[:30]
best_features_names = feature_names[best_features]
print(best_features_names)

```

```python
train[best_features_names].describe()
```

# KNeighborsClassifier

```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score


knn_1 = KNeighborsClassifier(n_neighbors=5)
knn_1.fit(X_train[best_features_names], y_train)
knn_2 = KNeighborsClassifier(n_neighbors=10)
knn_2.fit(X_train[best_features_names], y_train)
knn_3 = KNeighborsClassifier(n_neighbors=15)
knn_3.fit(X_train[best_features_names], y_train)
knn_4 = KNeighborsClassifier(n_neighbors=20)
knn_4.fit(X_train[best_features_names], y_train)
knn_5 = KNeighborsClassifier(n_neighbors=25)
knn_5.fit(X_train[best_features_names], y_train)

plt.figure(figsize=(20, 20))
predict_KNN_1 = knn_1.predict_proba(X_test[best_features_names])
predict_KNN_2 = knn_2.predict_proba(X_test[best_features_names])
predict_KNN_3 = knn_3.predict_proba(X_test[best_features_names])
predict_KNN_4 = knn_4.predict_proba(X_test[best_features_names])
predict_KNN_5 = knn_5.predict_proba(X_test[best_features_names])


fpr, tpr, thresholds = roc_curve(y_test, predict_KNN_1[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('KNeighborsClassifier_1',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_KNN_2[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('KNeighborsClassifier_2',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_KNN_3[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('KNeighborsClassifier_3',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_KNN_4[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('KNeighborsClassifier_4',auc(fpr, tpr)))

fpr, tpr, thresholds = roc_curve(y_test, predict_KNN_5[:,1])
plt.plot(fpr, tpr, label='%s ROC (area = %0.2f)' % ('KNeighborsClassifier_5',auc(fpr, tpr)))


plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.legend(loc=0, fontsize='small')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```
